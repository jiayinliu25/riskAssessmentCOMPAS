```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.preprocessing import OneHotEncoder

import warnings

warnings.filterwarnings('ignore', category=FutureWarning, module='sklearn')
```

```python

compas = pd.read_csv('data/compas-scores-two-years-violent.csv')

features = compas[['age', 'priors_count']]
target = compas['two_year_recid']

train_features, test_features, train_target, test_target = train_test_split(features, target, test_size=0.2, random_state=42)

# Define the neural network architecture
model = Sequential()
model.add(Dense(10, input_dim=2, activation='relu'))  # Input layer and first hidden layer
model.add(Dense(1, activation='sigmoid'))  # Output layer

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(train_features, train_target, epochs=100, batch_size=10, validation_split=0.1)

accuracy = model.evaluate(test_features, test_target)
```

```python
# Confusion matrix for NN with 2 features

# Generate predictions
nn_predictions = (model.predict(features) > 0.5).astype(int).flatten()

# Confusion matrix for the neural network model
conf_matrix_nn = confusion_matrix(compas['two_year_recid'], nn_predictions)

print("Confusion Matrix for Neural Network Model:")
print(conf_matrix_nn)

compas_binary_predictions = (compas['v_decile_score'] > 5).astype(int)

# Confusion matrix for COMPAS
conf_matrix_compas = confusion_matrix(compas['two_year_recid'], compas_binary_predictions)

print("Confusion Matrix for COMPAS Score:")
print(conf_matrix_compas)
```

```python

# Encode the categorical variables
encoder = OneHotEncoder(sparse=False)
categorical_features = encoder.fit_transform(compas[['sex', 'age_cat']])

categories = encoder.categories_
encoded_feature_names = [f"{col}_{cat}" for col, cats in zip(['sex', 'age_cat'], categories) for cat in cats]

encoded_df = pd.DataFrame(categorical_features, columns=encoded_feature_names)

# concatenate non-categorial and encoded categorical variables
compas_encoded = pd.concat([compas, encoded_df], axis=1)

feature_names = encoded_feature_names + ['juv_misd_count', 'juv_other_count', 'priors_count.1']
features = compas_encoded[feature_names]

target = compas['two_year_recid']

train_features, test_features, train_target, test_target = train_test_split(features, target, test_size=0.2, random_state=42)

# Define the neural network architecture
input_dim = train_features.shape[1]
model = Sequential()
model.add(Dense(10, input_dim=input_dim, activation='relu'))  # Adjusted input layer
model.add(Dense(1, activation='sigmoid'))  # Output layer

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(train_features, train_target, epochs=100, batch_size=10, validation_split=0.1)

accuracy = model.evaluate(test_features, test_target)
```

```python
# Confusion matrix for NN with 5 features

nn_predictions = (model.predict(features) > 0.5).astype(int).flatten()

conf_matrix_nn = confusion_matrix(compas['two_year_recid'], nn_predictions)

print("Confusion Matrix for Neural Network Model:")
print(conf_matrix_nn)

print("Confusion Matrix for COMPAS Score:")
print(conf_matrix_compas)

```

```python
# Load resampled data
resample_compas = pd.read_csv('combined_resample.csv')

resample_features = resample_compas[['age', 'priors_count']]
resample_target = resample_compas['two_year_recid']  

train_features, _, train_target, _ = train_test_split(resample_features, resample_target, test_size=0.2, random_state=42)

# Define the neural network architecture
model = Sequential()
model.add(Dense(10, input_dim=2, activation='relu'))  # 2 input features: 'age' and 'priors_count'
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(train_features, train_target, epochs=100, batch_size=10, validation_split=0.1)


compas_test_features = compas[['age', 'priors_count']]
compas_test_target = compas['two_year_recid'] 

accuracy = model.evaluate(compas_test_features, compas_test_target)
```

```python
# Confusion matrix for NN with 2 features and resampled data

compas_predictions = (model.predict(compas_test_features) > 0.5).astype(int).flatten()
conf_matrix_nn = confusion_matrix(compas_test_target, compas_predictions)

print("Confusion Matrix for Neural Network Model:")
print(conf_matrix_nn)

print("Confusion Matrix for COMPAS Score:")
print(conf_matrix_compas)
```

```python
# Encode categorical variables
encoder = OneHotEncoder(sparse=False)
new_categorical_features = encoder.fit_transform(resample_compas[['sex', 'age_cat']])

categories = encoder.categories_
new_encoded_feature_names = [f"{col}_{cat}" for col, cats in zip(['sex', 'age_cat'], categories) for cat in cats]

new_encoded_df = pd.DataFrame(new_categorical_features, columns=new_encoded_feature_names)

# concatenate
resample_compas_encoded = pd.concat([resample_compas, new_encoded_df], axis=1)

new_features = resample_compas_encoded[new_encoded_feature_names + ['juv_misd_count', 'juv_other_count', 'priors_count.1']]
new_target = resample_compas['two_year_recid']

train_features, _, train_target, _ = train_test_split(new_features, new_target, test_size=0.2, random_state=42)

# Define the neural network architecture
new_input_dim = train_features.shape[1]
model = Sequential()
model.add(Dense(10, input_dim=new_input_dim, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(train_features, train_target, epochs=100, batch_size=10, validation_split=0.1)

# Convert 'Male' to 1 and 'Female' to 0 in the 'compas' df
compas['sex'] = compas['sex'].map({'Male': 1, 'Female': 0})

compas_categorical_features = encoder.transform(compas[['sex', 'age_cat']])
compas_encoded_df = pd.DataFrame(compas_categorical_features, columns=new_encoded_feature_names)
compas_encoded = pd.concat([compas, compas_encoded_df], axis=1)
compas_features = compas_encoded[new_encoded_feature_names + ['juv_misd_count', 'juv_other_count', 'priors_count.1']]
compas_target = compas['two_year_recid']

# using original df for testing
original_test_features = compas_encoded[new_encoded_feature_names + ['juv_misd_count', 'juv_other_count', 'priors_count.1']]
original_test_target = compas['two_year_recid']

accuracy = model.evaluate(original_test_features, original_test_target)
```

```python
# Confusion matrix for NN with 5 features and resampled data

compas_predictions = (model.predict(compas_features) > 0.5).astype(int).flatten()

conf_matrix_compas_nn = confusion_matrix(compas['two_year_recid'], compas_predictions)

print("Confusion Matrix for Neural network model")
print(conf_matrix_compas_nn)

print("Confusion Matrix for COMPAS Score:")
print(conf_matrix_compas)
```
